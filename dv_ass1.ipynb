```python
 import pandas as pd
 import matplotlib.pyplot as plt
 import seaborn as sns
 import numpy as np
```

```python
df = pd.read_csv("country_wise_latest.csv")
df.head()
```

```python
 total_confirmed = df['Confirmed'].sum()
 total_deaths = df['Deaths'].sum()
 total_recovered = df['Recovered'].sum()
 total_active = df['Active'].sum()
```

```python
summary_df = pd.DataFrame({
    'Category' : ['Confirmed', 'Deaths', 'Recovered', 'Active'],
    'Total Cases' : [total_confirmed, total_deaths, total_recovered, total_active]
})

def format_number(num):
  if num >= 1_000_000:
   return f'{num / 1_000_000:.1f}M'
  elif num >= 1000:
   return f'{num / 1000:.1f}k'
  else:
   return str(num)

summary_df['Total Cases Formatted'] = summary_df['Total Cases'].apply(format_number)

plt.figure(figsize=(8, 6))
sns.barplot(x='Category', y='Total Cases', data=summary_df)
plt.title('Total Cases by Category')
plt.ylabel('Number of Cases')

for index, row in summary_df.iterrows():
   plt.text(index, row['Total Cases'], row['Total Cases Formatted'], color='black', ha="center", va='bottom')
plt.show()
```

```python
 sizes = [total_confirmed, total_deaths, total_recovered, total_active]
 labels = ['Confirmed', 'Deaths', 'Recovered', 'Active']

 plt.figure(figsize=(8, 8))
 plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=50)
 plt.title('Distribution of Cases')
 plt.show()
```

```python
 region_cases = df.groupby('WHO Region')[['Confirmed', 'Deaths', 'Recovered']].sum().reset_index()
 region_cases_melted = region_cases.melt(id_vars='WHO Region', var_name='Case Type', value_name='Total Cases')

 plt.figure(figsize=(12, 7))
 sns.barplot(x='WHO Region', y='Total Cases', hue='Case Type', data=region_cases_melted)
 plt.title('Total Cases by WHO Region')
 plt.ylabel('Number of Cases')
 plt.xlabel('WHO Region')
 plt.xticks(rotation=45, ha='right')
 plt.tight_layout()
 plt.show()
```

```python
sns.barplot(x='Deaths / 100 Cases', y='WHO Region', data=df, )
plt.title('Mortality Rate by WHO Region')
plt.show()
```

```python
 region_total_cases = region_cases_melted.groupby('WHO Region')['Total Cases'].sum().reset_index()

 plt.figure(figsize=(8, 8))
 plt.pie(region_total_cases['Total Cases'], labels=region_total_cases['WHO Region'], autopct='%1.1f%%', startangle=140)
 plt.title('Distribution of Total Cases by WHO Region')
 plt.show()
```

```python
 sns.barplot(x='Recovered / 100 Cases', y='WHO Region', data=df)
 plt.title('Recovered Rate by WHO Region')
 plt.show()
```

```python
 region_weekly_increase = df.groupby('WHO Region')['1 week change'].sum().reset_index()

 plt.figure(figsize=(12, 7))
 sns.barplot(x='WHO Region', y='1 week change', data=region_weekly_increase)
 plt.title('Region-wise Weekly Increase in Cases')
 plt.ylabel('Weekly Increase in Cases')
 plt.xlabel('WHO Region')
 plt.xticks(rotation=45, ha='right')
 plt.tight_layout()
 plt.show()
```

```python
top_10_recovered = df.nlargest(10, 'Recovered')

plt.figure(figsize=(12, 7))
sns.barplot(x='Recovered', y='Country/Region', data=top_10_recovered)
plt.title('Top 10 Countries by Recovered Cases')
plt.xlabel('Number of Recovered Cases')
plt.ylabel('Country/Region')
plt.tight_layout()
plt.show()
```

```python
%%writefile app.py


import streamlit as st
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Set the title of the Streamlit app
st.title("COVID-19 Data Storytelling App")

# Load the dataset
# Use @st.cache_data to cache the data loading operation for better performance
@st.cache_data
def load_data():
    # Read the csv file into a pandas DataFrame
    df = pd.read_csv("country_wise_latest.csv")
    # Perform some data cleaning (handle potential NaN missing values or duplicates if necessary)
    df.dropna(inplace=True)
    df.drop_duplicates(inplace=True)
    return df

# Load the data
df = load_data()

# Dataset Overview Section
st.header("Dataset Overview")
st.write("First 5 rows of the dataset:")
st.dataframe(df.head())

# Display the basic information about the dataset
st.write("Dataset Information:")
st.text(df.info())

# Display summary statistics
st.write("Summary Statistics:")
st.write(df.describe())

# Introduction Section
st.markdown("""
This app tells a story about the COVID-19 pandemic using data on confirmed cases, deaths, and recoveries by country.
It aims to visualize key trends, compare the impact across different countries and WHO regions, and highlight interesting patterns and anomalies in the data.
The data used in this app is from the provided "country_wise_latest.csv" dataset.
""")

st.header("Trends (Latest Data)")
st.markdown("""
This section visualizes the distribution of different case categories across countries based on the latest available data.
While this dataset does not contain time-series data for the trend lines, we can still observe the distribution of cases, deaths, and recoveries
at a single point in time across different countries or regions.
""")

# Widget for metric selection
metrics = st.selectbox(
    "Select Metric to Visualize:",
    ("Confirmed", "Deaths", "Recovered", "Active")
)

# This option will allow us to show trends over time directly
# However, as this dataset does not contain time-series data, we show the top 10 countries only

st.subheader("Top 10 Countries by {metrics} Cases")

# Get the top 10 countries for the selected metric
top_10_countries = df.nlargest(10, metrics)

# Create a bar chart to visualize the top 10 countries
plt.figure(figsize=(12,7))
sns.barplot(x=metrics, y='Country/Region', data=top_10_countries, palette='viridis')
plt.title(f"Top 10 Countries by {metrics} Cases - Latest Data")
plt.xlabel(metrics)
plt.ylabel('Country/Region')
plt.grid(axis='x', linestyle='--', alpha=0.6)
st.pyplot(plt)
plt.close()

# Conclusion summary section for the plot
st.markdown("""
This bar plot shows the top 10 countries with the highest number of {metrics} cases
based on the latest data in the dataset. It highlights the countries most affected
by this metric at the time the data was collected. The colors represent the different countries.
""")


# --- Comparisons Section ---
st.header("Comparisons")

st.markdown("""
This section allows you to compare the distribution of Confirmed, Deaths, and Recovered cases
across different WHO Regions. Select the regions you want to compare from the dropdown below.
""")

# Widget for selecting WHO Regions
# Allow the user to select multiple WHO regions for comparison
selected_regions = st.multiselect(
    'Select WHO Region(s) for Comparison',
    df['WHO Region'].unique(),
    default=df['WHO Region'].unique()  # Set a default selection to all regions
)

# Proceed only if at least one region is selected
if selected_regions:
    # Filter data for selected regions
    filtered_df = df[df['WHO Region'].isin(selected_regions)]

    # Group by WHO Region and sum cases for Confirmed, Deaths, and Recovered
    region_cases = filtered_df.groupby('WHO Region')[['Confirmed', 'Deaths', 'Recovered']].sum().reset_index()

    # Melt the DataFrame for plotting in a grouped bar chart format
    region_cases_melted = region_cases.melt(id_vars='WHO Region', var_name='Case Type', value_name='Total Cases')

    # Generate grouped bar plot to compare case types across selected regions
    plt.figure(figsize=(14, 8))  # Increase figure size for better readability with multiple regions
    # Use a different color palette for the grouped bar plot
    sns.barplot(x='WHO Region', y='Total Cases', hue='Case Type', data=region_cases_melted, palette='viridis')
    # Set plot title and labels with increased font size
    plt.title('Total Cases by WHO Region Comparison', fontsize=16)
    plt.xlabel('Number of Cases', fontsize=12)
    plt.ylabel('WHO Region', fontsize=12)
    # Rotate x-axis labels for better readability
    plt.xticks(rotation=45, ha='right')
    # Add grid axis y, linestyle='--', alpha=0.6)
    plt.grid(axis='y', linestyle='--', alpha=0.6)
    # Adjust layout to prevent labels from overlapping
    plt.tight_layout()

    # Display the plot
    st.pyplot(plt)
    # Close the figure
    plt.close()

    # add an explanatory caption for the plot
    st.markdown("""
    This grouped bar plot compares the total number of Confirmed, Deaths, and Recovered cases
  for the selected WHO Regions. It helps visualize the relative impact of the pandemic
across these different geographical areas. Each color represents a different case type (Confirmed, Deaths, Recovered).
""")
else:
    # Display a message if no regions are selected
    st.info("Please select at least one WHO Region to see the comparison.")

# --- Key Insights and Anomalies Section ---
st.header("Key Insights and Anomalies")

st.markdown("""
This section explores some interesting patterns, potential anomalies, and key insights
observed in the COVID-19 data. Visualizations are used to highlight these observations.
""")

st.subheader("Distribution of Death and Recovered Rates by WHO Region")

# Visualize the distribution of Deaths / 100 Cases by WHO Region using a box plot
# Improve plot aesthetics
plt.figure(figsize=(10, 6))
# Use a different color palette for the box plot
sns.boxplot(x='Deaths / 100 Cases', y='WHO Region', data=df, palette='plasma')
# Set plot title and labels with increased font size
plt.title('Distribution of Death Rate by WHO Region', fontsize=16)
plt.xlabel('Deaths per 100 Cases', fontsize=12)
plt.ylabel('WHO Region', fontsize=12)
# Add grid for readability on the x-axis
plt.grid(axis='x', linestyle='--', alpha=0.6)
# Adjust layout
plt.tight_layout()
# Display the plot
st.pyplot(plt)
# Close the figure
plt.close()

# Add an explanatory caption for the plot
st.markdown("""
This box plot shows the distribution of the death rate (Deaths per 100 Cases) across different WHO Regions.
It allows us to compare the variability and median death rates in each region, highlighting
regions with potentially higher or lower mortality rates relative to confirmed cases. Each box represents a WHO Region.
""")

# Visualize the distribution of Recovered / 100 Cases by WHO Region using a box plot
# Improve plot aesthetics
plt.figure(figsize=(10, 6))
# Use a different color palette for the box plot
sns.boxplot(x='Recovered / 100 Cases', y='WHO Region', data=df, palette='plasma')
# Set plot title and labels with increased font size
plt.title('Distribution of Recovered Rate by WHO Region', fontsize=16)
plt.xlabel('Recovered per 100 Cases', fontsize=12)
plt.ylabel('WHO Region', fontsize=12)
# Add grid for readability on the x-axis
plt.grid(axis='x', linestyle='--', alpha=0.6)
# Adjust layout
plt.tight_layout()
# Display the plot
st.pyplot(plt)
# Close the figure
plt.close()

# Add an explanatory caption for the plot
st.markdown("""
Similarly, this box plot illustrates the distribution of the recovered rate (Recovered per 100 Cases)
by WHO Region. It helps in understanding how recovery rates vary across different
regions and identifying potential differences in healthcare capacity or reporting standards. Each box represents a WHO Region.
""")

st.subheader("Relationship between Confirmed Cases and Deaths")

# Visualize the relationship between Confirmed and Deaths using a scatter plot
# Improve plot aesthetics
plt.figure(figsize=(10, 7))
# Use a specific color for the scatter points and set transparency
sns.scatterplot(x='Confirmed', y='Deaths', data=df, alpha=0.6, color='coral')
# Set plot title and labels with increased font size
plt.title('Relationship between Confirmed Cases and Deaths', fontsize=16)
# Update labels to indicate log scale
plt.xlabel('Confirmed Cases (Log Scale)', fontsize=12)
plt.ylabel('Deaths (Log Scale)', fontsize=12)
# Use log scale for both x and y axes due to wide range of values
plt.xscale('log')
plt.yscale('log')
# Add grid for readability
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
# Adjust layout
plt.tight_layout()
# Display the plot
st.pyplot(plt)
# Close the figure
plt.close()

# Title
st.title("Correlation Matrix of Numerical Features")

# Load dataset
@st.cache_data
def load_data():
    df = pd.read_csv("country_wise_latest.csv")
    df.dropna(inplace=True)
    return df

df = load_data()

# Introduction text
st.markdown("""
This heatmap visualizes the correlation between different numerical features in the dataset.
Correlation values range from -1 to +1, indicating the strength and direction of the linear relationship.
Values closer to 1 or -1 indicate a strong positive or negative correlation, respectively,
while values closer to 0 indicate a weak correlation.
""")

# Select only numerical columns for correlation
numerical_df = df.select_dtypes(include=['number'])

# Calculate correlation matrix
corr = numerical_df.corr()

# Plot heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(
    corr,
    annot=True,
    fmt=".2f",
    cmap="coolwarm",
    cbar=True,
    square=True,
    linewidths=0.5,
    linecolor='gray',
    annot_kws={"size":8}
)
plt.title("Correlation Matrix of Numerical Features", fontsize=16)
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()

st.pyplot(plt)
plt.close()

# Explanation text
st.markdown("""
The heatmap shows the pairwise correlation coefficients between numerical columns. For instance, a high positive correlation between **Confirmed** and **Deaths** suggests that as confirmed cases increase, deaths also tend to increase. This helps in understanding the relationships between different aspects of the pandemic data.
""")

# Conclusion section
st.header("Conclusion")

st.markdown("""
This app provided a glimpse into the global impact of the COVID-19 pandemic using the latest available country-wise data. We observed the overall distribution of confirmed cases, deaths, and recoveries, highlighting the significant burden of the pandemic worldwide.

Through regional comparisons, we saw how the pandemic’s impact varied across different WHO regions, with the Americas and Europe showing particularly high numbers of cases and deaths. Analysis of death and recovered rates by region revealed differences in outcomes, potentially influenced by factors such as healthcare systems, demographics, and public health measures.

The exploration of key insights and relationships, like the strong correlation between confirmed cases and deaths, further illuminated the severity of the virus and its consequences. While this dataset provides a snapshot, it effectively illustrates the scale and varied regional impact of the COVID-19 pandemic.
""")


```
